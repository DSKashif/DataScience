```python
# c.Creating columns for month, date, year using InvoiceDate Column
```

```python
data['year'] = data['InvoiceDate'].dt.year
data['month'] = data['InvoiceDate'].dt.month
data['date'] = data['InvoiceDate'].dt.day
data['hour'] = data['InvoiceDate'].dt.hour
data.head()
```

```python
# d.Creating columns for day of week we are adding + 1 to make monday = 1 , by default it is 0
```

```python
data['day'] = (data['InvoiceDate'].dt.dayofweek) + 1
data.head()
```

```python
data['Revenue'] = data['Quantity'] * data['UnitPrice']
data.head()
```
